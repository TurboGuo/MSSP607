```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

students_df = pd.read_csv('students.csv')
connections_df = pd.read_csv('connections.csv')
```

```python
G = nx.Graph()

# Add nodes with demographic attributes
for index, row in students_df.iterrows():
    G.add_node(row['Student_ID'], Age=row['Age'], Gender=row['Gender'], Major=row['Major'])

# Add edges representing student connections
for index, row in connections_df.iterrows():
    G.add_edge(row['Student_ID_1'], row['Student_ID_2'])

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

nx.draw(G, with_labels=True)
plt.show()
```
